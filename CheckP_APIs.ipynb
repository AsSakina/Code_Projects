```python
#Importing Libraries
#Install the requests and pandas libraries if you have not already done so.
import pandas as pd
import json
import requests
```

```python
#Use the requests library to send a request to the OMDb API to retrieve data about the top 10 movies.
import requests
api_url = "http://www.omdbapi.com/?apikey=d2a0a307&s=Murder&type=movie&page=1."
response = requests.get(api_url)
data = response.json()
data
```

```python
dataset = pd.DataFrame.from_dict(data['Search'])
dataset
```



---


- We can observe that **t** is more detailed than **s**

---
- In the following code, we create a list containing all the film titles in our dataset
- Then we create a new list film
- We perform an iteration with the for loop of our titles
- And we initialize article, which will contain all the names of the titles.
- We concatenate these with url
- We use a GET request that will be sent to a JSONPlaceholder
- The data in the response is in JSON format
- Then finally we use append to add all the elements in the list film initialised at debu


```python
title = list(dataset['Title'])
film = []
for movie in title:
  article = f"t={movie}&type=movie&page=1" # input article
  url = "http://www.omdbapi.com/?apikey=d2a0a307&"+article
  response=requests.get(url)
  data = response.json()
  film.append(data)
film
```



---


- now at the end we initialise dtf which will represent the conversion of our list into a dataframe


---



```python
dtf = pd.DataFrame(film, columns = ['Title', 'Year', 'Director', 'imdbRating'])
dtf
```
